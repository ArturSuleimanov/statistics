Линия регрессии, как и любая другая, задается уравнением вида y = b0 + b1 * x,
где b0 - интерсепт,
b1 - slope.
Искать параметры регрессии (b0 и b1), мы будем методом наименьших квадратов.
Мы подберем такие параметры, что сумма квадратов ошибок (остатков) была минимальна. Остатком является расстояние от точки до регрессионной прямой по оси Y.

b1 = (sdy / sdx) * corr,
где sdy и sdx - стандартные отклонения переменных x и y соответственно.

b0 = y_mean - b1 * x_mean, где
y_mean и x_mean - средние значения переменных

```python
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import math
```

```python
data = pd.read_csv("https://stepik.org/media/attachments/lesson/8086/states.csv")
```

```python
def cov(x, y):
    return sum(
        (x - np.mean(x)) * (y - np.mean(y))
    ) / len(x)

def corr(x, y):
    return cov(x, y) / (np.std(x) * np.std(y))
```

```python
cor = corr(data.hs_grad, data.poverty)
b1 = (np.std(data.poverty) / np.std(data.hs_grad)) * cor
b0 = np.mean(data.poverty) - b1 * np.mean(data.hs_grad)
print(b0, b1)
```

```python
fig, ax = plt.subplots(figsize = [15, 10])
ax.set_xlabel("hs_grad")
ax.set_ylabel("poverty")
ax.scatter(x = data.hs_grad, y = data.poverty)
x = np.array([i for i in range(int(min(data.hs_grad)), int(max(data.hs_grad)))])
ax.plot(x, b0 + x * b1, color = "r")  # cтроим линию регрессии
# На графике видно, что с ростом образования, снижается уровень бедности в штате.
```
