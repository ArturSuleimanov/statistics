Условия применения регрессионного анализа.
- Линейная взаимосвзязь X и Y.
- Нормальное распределение остатков.
- Гомоскедастичность - постоянная изменчивость остатков на всех уровнях независимой переменной.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import scipy.stats as ss
import pylab 
```

```python
data = pd.read_csv("http://d396qusza40orc.cloudfront.net/statistics/lec_resources/states.csv")
# metro_res говорит какой процент населения штата живет в столичной области
```

```python
fig, ax = plt.subplots(figsize = [15, 10])
ax.set_xlabel("hs_grad")
ax.set_ylabel("poverty")
ax.scatter(x = data.hs_grad, y = data.poverty)
# Видим, что взаимосвязь линейная
```

1. Сначала построим уравнение регрессии.
2. Далее найдём R^2 и гипотезы:
   H0: b1 = 0,
   H1: b1 != 0 (см. Гипотеза о значимости взаимосвязи и коэффициент детерминации).
3. Посмотрим, как основываясь на данных независимой переменной можно предсказать, чему равна зависимая переменная.

```python
def cov(x, y):
    return sum(
        (x - np.mean(x)) * (y - np.mean(y))
    ) / len(x)

def corr(x, y):
    return cov(x, y) / (np.std(x) * np.std(y))

cor = corr(data.hs_grad, data.poverty)    # Коэффициент корреляции
b1 = (np.std(data.poverty) / np.std(data.hs_grad)) * cor
b0 = np.mean(data.poverty) - b1 * np.mean(data.hs_grad)
print(b0, b1)
```

```python
# или то же самое с помощью подключаемых библиотек
model_linear_ols = smf.ols(formula='poverty ~ hs_grad', data=data)
result_linear_ols = model_linear_ols.fit()
print(result_linear_ols.summary())
```

Мы получили b1 = -0.6212, b0 = 64.7810.

Получим уравнение: y = 64.7810 - 0.6212 * hs_grad

```python
fig, ax = plt.subplots(figsize = [15, 10])
ax.set_xlabel("hs_grad")
ax.set_ylabel("poverty")
ax.scatter(x = data.hs_grad, y = data.poverty)
x = np.array([i for i in range(int(min(data.hs_grad)), int(max(data.hs_grad)))])
ax.plot(x, b0 + x * b1, color = "r")  # cтроим линию регрессии
# На графике видно, что с ростом образования, снижается уровень бедности в штате.
```

```python
# Для проверки гипотезы о том, что b1 = 0, мы использовали T-тест, его результаты
# можно увидеть в таблице
"""
p_value < 0.05 => значит мы имеем статистически значимые отличия и можем отклонить
нулевую гипотезу.
b0 = 64.7810 говорит нам о том, что если предположить, что есть штат, в котором
процент людей с образованием равен 0, мы получим 65% людей проживающих за чертой
бедности.
R^2 = 0.558 означает, что 56 процентов изменчивости нашей зависимой переменной
обусловлено нашей регрессионной моделью.
"""
data["hs_grad_residuals"] = data.poverty - b0 - b1*data.hs_grad    # находим остатки
```

```python
ss.probplot(data.hs_grad_residuals, dist="norm", plot=pylab)
pylab.show()    # Строим qq-plot для остатков
# и видим, что распеделение относительно нормальное
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("hs_grad")
ax.set_ylabel("hs_grad_residuals")
ax.scatter(x = data.hs_grad, y = data.hs_grad_residuals)
x = np.array([i for i in range(int(min(data.hs_grad)), int(max(data.hs_grad)))])
ax.plot(x, [np.mean(data.hs_grad_residuals) for i in range(len(x))], color = "r")
# смотрим на распределение остатков вокруг среднего
# скаттер плот с остатками показывает гетероскедастичность остатков
```
