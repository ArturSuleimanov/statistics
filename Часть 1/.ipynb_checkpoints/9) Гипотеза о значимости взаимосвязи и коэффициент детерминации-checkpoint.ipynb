Гипотеза о значимости взаимосвязи (угла наклона регрессионной прямой).

Если у наших переменных нет значимой взаимосвязи, то коэффициент slope или b1 будет равен 0. Следовательно регрессионная прямая пройдёт параллельно оси Ox.

Поэтому мы можем сформулировать нулевую гипотезу, согласно которой при отсутствии значимой взаимосвязи между переменными, при многочисленном повторении выборки коэффициент b1 распределится нормально вокруг среднего равного 0.
H0: b1 = 0
H1: b1 != 0

Тогда Т-критерий будет равен (b1 - 0) / se, где 0 - это среднее значение коэффициента b1 по нашей нулевой гипотезе. 

Т критерий проверяет гипотезу о том, что Т - критерий значимо отличен от нуля.

Число степеней свободы определится по формуле N - 2, где N - число наблюдений.

Коэффициент детерминации (R^2) - доля дисперсии зависимой переменной Y, обуславлемая нашей регрессионной моделью.

R^2 = 1 - SSres / SStotal, где
SSres - сумма квадратов остатков (расстояние от нашего наблюдения до регрессионной прямой возведенное в квадрат),
SStotal - сумма растояний от нашего наблюдения до среднего значения по переменной.
