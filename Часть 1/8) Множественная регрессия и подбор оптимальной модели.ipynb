Мультиколинеарность - очень сильная взаимосвязь независимых переменных.

Мультиколинеарность может испортить качество нашей модели и сильно снизит R^2.

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import numpy as np
```

```python
data = pd.read_csv("http://d396qusza40orc.cloudfront.net/statistics/lec_resources/states.csv")
```

```python
data.corr()
# тут мы видим, что переменная female_house сильно коррелирует со многими другими
# переменными, поэтому лучше убрать ее из нашей модели
"""
подбор оптимальной модели заключается в поочередном удалении каждой переменной
из нашей модели, и выбор из этих моделей модели с наилучшим исправленным R^2.
Далее из полученной модели снова попробуем удалить оставшиеся независимые переменные.
"""
# оптимальная модель будет выглядеть так
model_linear_ols = smf.ols(formula='poverty ~ white + metro_res + hs_grad', data=data)
result_linear_ols = model_linear_ols.fit()
linear_params = result_linear_ols.params
linear_params
```

```python
model_linear_ols = smf.ols(formula='poverty ~ hs_grad + white', data=data)
result_linear_ols = model_linear_ols.fit()
```

```python
linear_params = result_linear_ols.params
print(linear_params)
print("R^2", result_linear_ols.rsquared)    # вот такая часть изменчивости бедности объясняется этой моделью
print("Исправленный R^2", result_linear_ols.rsquared_adj) # расчитывается с учетом того, что в нашей модели несколько предикторов



# эти коэффициеты показывают, что при неизменных значения white, процент
# людей проживающих в бедности упадет на 0.030888 процента
```

```python
# Построим трехмерный scatter plot
fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')
xx, yy = np.meshgrid(range(int(min(data.hs_grad)),int(max(data.hs_grad))),
                     range(int(min(data.white)),int(max(data.white))))
z = (linear_params[0] + linear_params[1] * xx + linear_params[2] * yy)
ax.plot_surface(xx, yy, z, alpha=0.25)
ax.set_xlabel("hs_grad")
ax.set_ylabel("white")
ax.set_zlabel("poverty")
ax.scatter(data.hs_grad, data.white, data.poverty)
plt.show()
"""
На графике видно, что с ростом процента белого населения и ростом процента людей,
закончивших среднюю школу, падает уровень бедности.
"""
```
