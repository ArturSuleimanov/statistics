```python
import numpy as np
import pandas as pd
import seaborn as sns
import scipy.stats as stats
import math
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import statsmodels.api as sm
```

```python
data = pd.read_csv("https://stepic.org/media/attachments/course/524/train.csv")
```

Теперь мы имеем 2 предиктора: Pclass - класс билета, Sex - пол. И мы попробуем предсказать выжил человек или нет, исходя из пола, из класса билета и взаимодействия этих 2-х предикторов. И даже увидим как меняется вероятность выживания разных полов в зависимости от класса их билета.

```python
# Преобразуем наши данные
data.loc[data["Pclass"] == 1, "Pclass"] = "FirstClass"
data.loc[data["Pclass"] == 2, "Pclass"] = "SecondClass"
data.loc[data["Pclass"] == 3, "Pclass"] = "ThirdClass"
data.loc[data["Survived"] == 1, "Survived"] = "Yes"
data.loc[data["Survived"] == 0, "Survived"] = "No"
```

```python
plt.rcParams['figure.figsize']=(12, 8) # ура, установили размер
mosaic(data, ['Survived', 'Sex', 'Pclass'], gap = 0.02, title='Survived from Titanic')
plt.show()
```

```python
# построим логистическую модель
model = sm.formula.glm("Survived ~ Sex * Pclass",
                       family=sm.families.Binomial(), data=data).fit()
model.summary()
```

В полученных данных мы видим только мужчин, а из классов видим только 2 и 3,
соответственно в Intercept отражена женщина в 1 классе.

```python
# Построим трехмерную таблицу, чтобы посмотреть выжили пассажиры или нет, в
# зависимости от пола и класса билета.
data.groupby(["Sex", "Pclass", "Survived"]).count()["PassengerId"]
```

Теперь расчитаем вручную.

```python
# В Intercept поместим вероятность выжить для женщин из 1 класса (самый безопасный билет)
female_p1_odds = 91 / 3
Intercept = math.log(female_p1_odds)

# Коэффициент Sex[T.male] - это логарифм отношения шансов выжить для мужчин в 1 классе
# к шансам выжить для женщин в 1 классе
male_p1_odds = 45 / 77
sex_male = math.log(male_p1_odds / female_p1_odds)

# Коэффициент Pclass[T.SecondClass] - это логарифм отношения шансов выжить для женщин во втором классе к шансам
# выжить для женщин в 1 классе
# Мы также увидим, что при переходе женщины во 2 класс, ее шансы выжить статистически значимо не меняются
female_p2_odds = 70 / 6
class_2 = math.log(female_p2_odds / female_p1_odds)

# Коэффициент Pclass[T.ThirdClass] - это логарифм отношения шансов выжить для женщин в третьем классе к шансам
# выжить для женщин в 1 классе
female_p3_odds = 72 / 72
class_3 = math.log(female_p3_odds / female_p1_odds)

# Коэффициент Sex[T.male]:Pclass[T.SecondClass] - можно понять исходя из формулы!
male_p2_odds = 17 / 91    # Шансы выжить для мужчин во 2 классе
# Также мы увидим, что шансы выжить для мужчин в 1 классе и во 2 классе статистически значимо от 0 не отличаются,
# т.к. пропорция выживших мужчин и женщин приблизительно одинакова и в 1 и во 2 классе
male_class_2 = math.log((male_p2_odds / female_p2_odds) / (male_p1_odds / female_p1_odds))

# Коэффициент Sex[T.male]:Pclass[T.ThirdClass] - можно понять исходя из формулы!
male_p3_odds = 47 / 300    # Шансы выжить для мужчин во 2 классе
# Здесь различия будут статистически значимы, т.к. в 3 классе шансы выжить для женщин существенно ниже, чем в 1 классе
male_class_3 = math.log((male_p3_odds / female_p3_odds) / (male_p1_odds / female_p1_odds))
```

Итерпретация:

Допустим мы получили новое наблюдение в нашей выборке - это мужчина в 3 классе. Сначала мы ставим значение Intercept, независимо от того, кем является этот человек. Понижаем шансы испытуемого т.к. он мужчина (sex_male = 1). Он не во втором классе (2_class = 0). Он в 3 классе, понижаем его шансы выжить, исходя и того, что он в 3 классе (3_class = 1). Далее sex_male * 2_class = 0, т.к. 2_class = 0. В 3 классе отличие между мужчинами и женщинами не такое сильное, поэтому штраф, который был наложен за то, что испытуемый мужчина, мы чуть-чуть поднимем, потому что 3_class = 1.
